```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

## Bivariate Plots

### Regression Plot

```python
plt.figure(figsize=(8, 5))
sns.regplot(x="total_bill", y="tip", data=tips, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.savefig("reg_plot.png")
plt.show()
```

### Joint plot

```python
sns.set(style="white")
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter")
plt.show()
```

```python
import warnings as wr
wr.filterwarnings("ignore")
```

```python
sns.set(style="white")
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde")
plt.savefig("joint_plt.png")
plt.show()
```

### Hexbin Plot

```python
plt.figure(figsize=(8, 5))
sns.jointplot(x="total_bill", y="tip",kind='hex', data=tips, gridsize=15, cmap="Blues")
plt.savefig("hexbin_plot.png")
```

## MultiVariate Plot

### Using Parameters

```python
plt.figure(figsize=(10, 6))
sns.violinplot(
    x="day",         # x-axis: Days of the week (categorical)
    y="total_bill",  # y-axis: Total bill amount (numerical)
    data=tips, 
    hue="sex", # Color by gender (categorical)
    palette="Set1",  # Color palette
    split=True       # Split violins by hue categories
)

plt.legend(title = "gender")
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x="total_bill", 
    y="tip", 
    data=tips, 
    hue="day",      
    size="tip", #for better undrestanding
    sizes=(20, 200), 
    palette="Set1" 
)
```

## Relational Plot

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="time", style="size", size="tip", palette="Set1", height=6)
plt.title("Relational Scatter Plot for Tips Dataset")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.savefig('relplot_with_mv.png')
plt.show()
```

```python
g = sns.relplot(
    x="total_bill",
    y="tip",
    data=tips,
    hue="time",
    style = 'sex',
    col="day",  # Separate plots by day (columns)
    row="sex",  # Separate plots by gender (rows)
    palette="Set1",
    height=3,  # Height of each subplot
    aspect=1.2  # Aspect ratio of each subplot
)
plt.savefig('relplot_col_row.png')
```

## Facet Grid

```python
g = sns.FacetGrid(tips,row = 'day', col="sex", height=4, aspect=1.2)
g.map(sns.histplot, "total_bill", kde=True)
plt.savefig("faceg_grid.png")
```

## Pair Plot

```python
sns.get_dataset_names()
```

```python
iris = sns.load_dataset("iris")
sns.set(style="ticks")
sns.pairplot(iris,diag_kind='hist', hue="species", markers=["o", "s", "D"])
plt.savefig('pair_plot.png')
plt.show()
```

```python
g = sns.PairGrid(iris, hue="species")
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot, kde_kws={"color": "k"})
g.map_lower(sns.kdeplot)
g.add_legend()
plt.savefig('pair_grid.png')
plt.show()
```

##  Matrix Plots


### Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset("titanic")

# Drop non-numeric columns to calculate the correlation matrix
correlation_matrix = titanic_data.select_dtypes(include=['number']).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Titanic Dataset")  # Corrected title
plt.savefig('corr_matrix.png')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Titanic dataset
data = sns.load_dataset("titanic")

# Select only numeric columns
numeric_data = data.select_dtypes(include=[np.number])

# Create a Cluster Map of the correlation matrix
sns.clustermap(numeric_data.corr(), annot=True, cmap="coolwarm", figsize=(10, 8))
plt.title("Cluster Map of Titanic Dataset Correlation (Numeric Features)")
plt.tight_layout()
plt.savefig('cluster_map.png')
plt.show()
```
