```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('fifa_players.csv')
df
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['age'][:20], df['overall_rating'][:20], marker='o', color='b')
plt.title('Age vs Overall Rating (First 20 Players)')
plt.xlabel('Age')
plt.ylabel('Overall Rating')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
df['nationality'].value_counts().head(10).plot(kind='bar', color='orange')
plt.title('Top 10 Nationalities by Player Count')
plt.xlabel('Nationality')
plt.ylabel('Number of Players')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['age'], bins=15, color='green', edgecolor='black')
plt.title('Distribution of Players\' Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['height_cm'], df['weight_kgs'], alpha=0.1, color='purple')
plt.title('Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
df['nationality'].value_counts().head(10).plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Top 5 Nationalities')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df['overall_rating'], vert=False)
plt.title('Box Plot of Overall Ratings')
plt.xlabel('Overall Rating')
plt.show()
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.hist(df['age'], bins=15, color='lightblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(df['overall_rating'], bins=15, color='salmon', edgecolor='black')
plt.title('Overall Rating Distribution')
plt.xlabel('Overall Rating')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
